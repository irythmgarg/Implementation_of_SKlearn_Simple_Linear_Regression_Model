# **IMPLEMENTATION OF SIMPLE LINEAR REGRESSION**

```python
import pandas as pd # import pandas
import numpy as np #import numpy
```

```python
df=pd.read_csv('/content/placement 2.csv'); # dataset of cgpa and placement
df.head();
```

```python
x=df.iloc[:,0].values;  # cgpa column
y=df.iloc[:,1].values;  # placement column
```

```python
df.shape # number of row and columns
```

```python
from sklearn.model_selection import train_test_split # spliting of dataset into training and testig part
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
y_train
```

```python
from sklearn.linear_model import LinearRegression #  inbuilt model  scikit learn class
lr=LinearRegression() # object of class
x_train_pd=pd.DataFrame(x_train);
x_test_pd=pd.DataFrame(x_test);
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test_pd); # prediction of test data
```

```python
pd.DataFrame(y_pred).head() # result of prediction
```

# This the class made by us to understand working of inbuilt scikit learn Linear regression class

```python
class OurLinearRegression :
  def __init__(self):
    self.m=None # slope of the minimum error line
    self.b=None; # intercept of the line
  def fit(self,x_train,y_train):
    num=0;
    den=0;
    for i in range(x_train.shape[0]):
       num+=(x_train[i]-x_train.mean())*(y_train[i]-y_train.mean());
       den+=(x_train[i]-x_train.mean())**2;
    self.m=num/den;
    self.b=y_train.mean()-self.m*x_train.mean();
  def predict(self,x_test):
    return self.m*x_test+self.b;
```

```python
mylr=OurLinearRegression() # our model object
mylr.fit(x_train,y_train)
```

```python
y_test #actual values
```

```python
mylr.predict(x_test) # our prediction
```
